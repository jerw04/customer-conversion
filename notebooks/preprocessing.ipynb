```python
import pandas as pd

url = "https://raw.githubusercontent.com/jerw04/customer-conversion/refs/heads/main/data/e-shop%20clothing%202008.csv"
uci = pd.read_csv(url, sep=";")

print("Shape:", uci.shape)
print("Columns:", uci.columns.tolist())

# First 5 rows
display(uci.head())

# Info
print("\nInfo:")
print(uci.info())

# Unique counts per column
print("\n--- Unique counts per column ---")
for col in uci.columns:
    print(f"{col}: {uci[col].nunique()}")
```

```python
uci = uci.rename(columns={
    "session ID": "session_id",
    "page 1 (main category)": "page1_main_category",
    "page 2 (clothing model)": "page2_clothing_model",
    "model photography": "model_photography",
    "price 2": "price_2"
})

print("Columns after renaming:", uci.columns.tolist())
uci.head()
```

```python
# Derive a simple target: mark a session as purchased if it has any "sale" page (category = 4)
purchase_sessions = (
    uci.groupby("session_id")["page1_main_category"]
       .apply(lambda x: 1 if 4 in x.values else 0)
)

# Merge back into main dataframe
uci = uci.merge(purchase_sessions, on="session_id", how="left")
uci = uci.rename(columns={"page1_main_category_y": "converted", "page1_main_category_x": "page1_main_category"})

print("Unique values in target (converted):", uci["converted"].unique())
print(uci[["session_id","page1_main_category","converted"]].head(20))
```

```python
# Aggregate features per session
session_df = uci.groupby("session_id").agg({
    "month": "first",
    "day": "first",
    "country": "first",
    "page1_main_category": lambda x: x.mode()[0],  # most common category in session
    "page2_clothing_model": lambda x: x.nunique(), # number of distinct products viewed
    "colour": lambda x: x.nunique(),               # diversity of colors seen
    "location": lambda x: x.nunique(),             # positions clicked
    "model_photography": lambda x: x.mode()[0],
    "price": ["mean","max"],                       # avg and max price seen
    "price_2": "mean",
    "page": "max",                                 # last page reached
    "converted": "max"                             # target
})

# Flatten column names
session_df.columns = ["_".join([col for col in tup if col]) for tup in session_df.columns]
session_df = session_df.reset_index()

print("Shape:", session_df.shape)
print("Columns:", session_df.columns.tolist())
session_df.head()
```
